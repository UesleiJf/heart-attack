```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns
import plotly.express as px
import plotly.figure_factory as ff
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('data/heart.csv')
data.head(5)
```

```python
heart=Counter(data['output'])

classes=[]
count=[]   #list to store no of labels of each class

for i in heart.keys():
    classes.append(i)
    count.append(heart[i])
    
colors = ["#7d7577", "#4c5500"]
labels = ['Tiveram Ataque', 'Não tiveram Ataque']
plt.pie(count, shadow = True, labels=labels, colors = colors, startangle=80, autopct='%1.1f%%', textprops={'color':"w"})
plt.axis('equal')
plt.tight_layout()
plt.title("Porcentagem de pessoas com ataque cardíaco no conjunto de dados", fontsize=15, color='w')
plt.show()
```
